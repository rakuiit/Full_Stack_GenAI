```python
folder_path='/Users/rks/Documents/projects/Full_Stack_GenAI/data'
story=[]
for file in os.listdir(folder_path):
    with open(os.path.join(folder_path,file),encoding="unicode_escape") as f:
        corpus=f.read()
        raw_sent=sent_tokenize(corpus)
        for i in raw_sent:
            story.append(simple_preprocess(i))
```

```python
story[0]
```

```python
## sg=1 for skip-gram and sg=0 for cbow
custom_model=gensim.models.Word2Vec(
    window=10,
    min_count=2,
    vector_size=150,


)
```

```python
custom_model.build_vocab(story)
```

```python
custom_model.corpus_count
```

```python
 custom_model.train(story,total_examples=custom_model.corpus_count,epochs=5)
```
